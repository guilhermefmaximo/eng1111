```python
import pandas as pd
from google.colab import drive

drive.mount('/drive')
data = pd.read_csv("/drive/My Drive/class_heart_disease.csv")

print(data.info())

display(data)
```

```python
data.dropna(subset = ["Age"], inplace = True)

#DATASHEET NAO POSSUI VALORES NULOS#

idade_media = data["Age"].mean()
print(idade_media)
```

```python
#removendo idades 10 anos ou mais diferente da media

remov = data[(abs(data['Age'] - idade_media) >= 10 )].index
ndata = data.copy()
ndata.drop(remov, inplace = True)
display(ndata)
```

```python
c1 = ndata["Age"]
c2 = ndata["Sex"]
c3 = ndata["Chol"]
c4 = ndata["Slope"]
c5 = ndata["Num"]

data3 = pd.concat([c1,c2,c3,c4,c5],axis = 1)
display(data3)
```

```python
import random

lista_idade = c1.tolist()
lista_chol = c3.tolist()

novo = list()

for i, el in enumerate(lista_idade):
  n = (el + lista_chol[i]) * random.randint(0,1)
  novo.append(n)

data3.insert(5,"NOVO",novo)
display(data3)
```

```python
import numpy as np

embaralhado = data3.reindex(np.random.permutation(data3.index))
display(embaralhado)
```

```python
#idade crescente

teste = embaralhado.sort_values(by='Age', ascending=True)
display(teste)
```

```python
#NAO TINHA COLUNA DE STRING

group = teste.groupby(["Age"])
print(group.size().reset_index(name='quant'))
```

```python
#dobrando as idades com map

teste["Age"]=teste["Age"].map(lambda x: x*2)
display(teste)
```

```python
#mandando pro drive

teste.to_csv("/drive/My Drive/teste.csv")
teste.to_json("/drive/My Drive/teste.json")
```
